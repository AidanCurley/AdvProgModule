```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("PeoplesFavourites-1.csv")
df
```

```python
# A number of different symbols (including space) have been used for null values, select and apply a consistant approach to their representation.
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df.replace(['0', 0, 'none', None, 'null'], np.nan, inplace=True)
df.fillna(method='ffill')
df
```

```python
# Remove any customer data that has missing data in any of the following first name, last name and email.
columns = ['Email', 'First Name', 'Last Name']
for column in columns:
    df = df[df[column].notna()]
df
```

```python
# Anonymize the data by removing the customer names (first and last name) replacing with a unique identifier.
df.columns
labels = ['First Name', 'Last Name']
df.drop(columns=labels, inplace= True)
df['unique_id'] = df.index
df.head()
```

```python
# Once complete export the data to a new file
df.to_csv('Activity3_results.csv', index=False)
```
