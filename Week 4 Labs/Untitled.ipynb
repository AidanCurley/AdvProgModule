```python
import pandas as pd

df = pd.read_csv("SalesData.csv", index_col=0)

print("df before:\n", df)

df = df.apply(lambda row : row.apply(lambda x : (x - row.mean())/row.mean())) 

print("df after:\n", df)

min_store, min_month = df.stack().idxmin()

max_store, max_month = df.stack().idxmax()

print("\n1. Which store had the largest and smallest increase and in which"+
      "months did they occur?\n")

print("\nlargest:\n",  df.loc[[max_store], [max_month]])

print("\nsmallest:\n",  df.loc[[min_store], [min_month]])


print("\n2. Across all the stores which months showed the smallest average"+
      "increase and the largest average increase?\n")

mean_pct_change = df.mean(axis = 1)

min_idx = mean_pct_change.idxmin()

min_value = mean_pct_change.loc[mean_pct_change.idxmin()]

max_idx = mean_pct_change.idxmax()

max_value = mean_pct_change.loc[mean_pct_change.idxmax()]

print("\nlargest:\n", max_idx, ":", max_value)

print("\nsmallest:\n", min_idx, ":",min_value)
```
